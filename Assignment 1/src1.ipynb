```python
#Alan Uthuppan

import pymnet as pmn
import networkx as nx
import matplotlib.pyplot as plt
import csv
import collections
import statistics
from itertools import combinations
```

```python
#Problem 1

multiNet = pmn.MultiplexNetwork(couplings="none")

multiNet.add_layer("Data Mining/Machine Learning")
multiNet.add_layer("Python/R")
multiNet.add_layer("Graphs/Statistics")

with open("Assignment_1_Data.csv", newline='') as file:
    datamining = list()
    python = list()
    graphs = list()
    
    data = csv.reader(file, delimiter=',', quotechar='|')
    next(data)
    for row in data:
        student = row[1]
        multiNet.add_node(student[8:])
        if row[2] == "Yes":
            datamining.append(student[8:])
        if row[3] == "Yes":
            python.append(student[8:])
        if row[4] == "Yes":
            graphs.append(student[8:])
            
for i in list(combinations(datamining, 2)):
    multiNet[i[0], i[1], "Data Mining/Machine Learning", "Data Mining/Machine Learning"] = 1
for i in list(combinations(python, 2)):
    multiNet[i[0], i[1], "Python/R", "Python/R"] = 1
for i in list(combinations(graphs, 2)):
    multiNet[i[0], i[1], "Graphs/Statistics", "Graphs/Statistics"] = 1

fig = pmn.draw(multiNet, layout='circular', autoscale=True, show=True) 
fig.savefig("Problem 1 Output.png")
```

```python
#Problem 2

G = nx.Graph()

with open("Assignment_1_Data.csv", newline='') as file:
    datamining = list()
    python = list()
    graphs = list()
    
    data = csv.reader(file, delimiter=',', quotechar='|')
    next(data)
    for row in data:
        student = row[1]
        G.add_node(student[8:])
        if row[2] == "Yes":
            datamining.append(student[8:])
        if row[3] == "Yes":
            python.append(student[8:])
        if row[4] == "Yes":
            graphs.append(student[8:])

for i in list(combinations(datamining, 2)):
    if G.has_edge(i[0], i[1]) == False:
        G.add_edge(i[0], i[1], weight=1)
    else:
        G.add_edge(i[0], i[1], weight=G[i[0]][i[1]]['weight']+1)
        
for i in list(combinations(python, 2)):
    if G.has_edge(i[0], i[1]) == False:
        G.add_edge(i[0], i[1], weight=1)
    else:
        G.add_edge(i[0], i[1], weight=G[i[0]][i[1]]['weight']+1)
        
for i in list(combinations(graphs, 2)):
    if G.has_edge(i[0], i[1]) == False:
        G.add_edge(i[0], i[1], weight=1)
    else:
        G.add_edge(i[0], i[1], weight=G[i[0]][i[1]]['weight']+1)
        
pos = nx.spring_layout(G)
nx.draw_networkx(G, with_labels=True, pos=pos)
plt.savefig("Problem 2 Output.png")            
```

```python
#Problem 3

G = nx.Graph()

with open("Assignment_1_Data.csv", newline='') as file:
    datamining = list()
    python = list()
    graphs = list()
    
    data = csv.reader(file, delimiter=',', quotechar='|')
    next(data)
    for row in data:
        student = row[1]
        G.add_node(student[8:], bipartite=0)
        if row[2] == "Yes":
            datamining.append(student[8:])
        if row[3] == "Yes":
            python.append(student[8:])
        if row[4] == "Yes":
            graphs.append(student[8:])


nodes = ["DM", "PY", "GS"]
G.add_nodes_from(nodes, bipartite=1)

for i in datamining:
    G.add_edge(i, nodes[0])
for i in python:
    G.add_edge(i, nodes[1])
for i in graphs:
    G.add_edge(i, nodes[2])

pos = nx.bipartite_layout(G, nodes, align="horizontal")    
nx.draw_networkx(G, pos=pos)
plt.savefig("Problem 3 Output.png") 
```
